```python
"""
for i in range(10):
    cap = cv2.VideoCapture(i)
    cap.set(cv2.CAP_PROP_FPS, 5)
    print(cap.get(3))
    #if cap.isOpened()==False:
        #print(i,"error")
    ret,frame = cap.read()
    if ret==True:
        print(i)
    cap.release()
"""
capleft = cv2.VideoCapture(0)#left image 
capright= cv2.VideoCapture(2)#right image
SIZE = (640,480)
while True:
    ret,framel = capleft.read()
    ret,framer = capright.read()
    print(framel.shape)
    imagel = cv2.resize(framel,SIZE)
    imager = cv2.resize(framer,SIZE)
    #image =frame
    #SIZE = image.shape
    ctrx = round(SIZE[1]/2)
    ctry = round(SIZE[0]/2)
    line_v = [(ctry,0),(ctry,SIZE[1])] #drawing the line vertically
    line_h_2 = [(0,ctrx+round(ctrx/2)),(SIZE[0],ctrx+round(ctrx/2))] #drawing the line horizontally at the 4th layer
    line_h = [(0,ctrx),(SIZE[0],ctrx)] #drawing the line horizontally
    #for left image
    cv2.line(imagel,line_v[0],line_v[1],(0,255,0),2)
    cv2.line(imagel,line_h[0],line_h[1],(0,255,0),2)
    cv2.circle(imagel,(ctry,ctrx),50,(0,255,0),2)
    #for right image
    cv2.line(imager,line_v[0],line_v[1],(0,255,0),2)
    cv2.line(imager,line_h_2[0],line_h_2[1],(0,255,0),2)
    cv2.line(imager,line_h[0],line_h[1],(0,255,0),2)
    cv2.circle(imager,(ctry,ctrx),50,(0,255,0),2)
    #print(image.shape)
    cv2.imshow("left_image",imagel)
    cv2.imshow("right_image",imager)
    k = cv2.waitKey(5)
    if k & 0xFF == ord("q"): # Exit condition
        break
cv2.destroyAllWindows()
capleft.release()
capright.release()
```
